Raw dataset was grouped by ‘utrip_id’ so, cities are ordered and grouped within the
same trip. 

Then the ordered cities are paired like (departure city, arrival city), for instance, if a
man goes through A, B, anc C, they are paired like (A,B), and (B,C). 

The same paired rows were counted.

So, final transformed data have three columns ‘departure_city’, ‘arrvial_city’, and
‘count’.

In this dataset, the total number of unique paired cities are more than three hundred
thousand rows.

```python
testting=train_set
#group by utrip_id
dfs=list(testting.groupby(testting['utrip_id']))
```

```python
len(dfs), type(dfs)
```

```python
dfs
```

```python
newlist = []
for i in range(len(dfs)):
    x=dfs[i][1]['city_id']
    x = [x for x in x if isinstance(x, (int,float))]

    newlist.append(x)
```

```python
newlist
```

```python
unique=[]
for i in range(len(newlist)):
    x=newlist[i]
    for j in x:
        if j not in unique:
            unique.append(j) 
```

```python
test_list=[]
for i in range(len(newlist)-1):
    test=newlist[i]
    for j in range(len(test)-1):
        test_list.append((test[j],test[j+1]))
        #print(test[j],test[j+1])
        
```

```python
test_list
```

```python
number_of_pairs = collections.defaultdict(int)
for elem in test_list:
    number_of_pairs[elem] += 1
number_of_pairs
```

```python
#print(number_of_pairs)
```

```python
#
departure_city = collections.defaultdict(int)
arrival_city = collections.defaultdict(int)

for departure,arrival in test_list:
    departure_city[departure] += 1
    arrival_city[arrival] += 1
```

```python
df = pd.DataFrame(number_of_pairs.items(), columns=['paired_cities','count'])
```

```python
df[['departure_city', 'arrival_city']] = pd.DataFrame(df['paired_cities'].tolist(), index=df.index)
df = df[['departure_city', 'arrival_city', 'count']]
```

```python
df
```

```python
df.to_csv('converted_dataset.csv', encoding='utf-8', index=False)
```
