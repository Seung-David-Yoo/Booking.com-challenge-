```python
import numpy as np
import pandas as pd
import collections
```

```python
train_set = pd.read_csv('train_set.csv').sort_values(by=['utrip_id','checkin'])
train_set.groupby(by=['utrip_id'])
train_set.head(20)
```

For lift based bayesian theorem, another dataset was generated. From the raw data,
after grouping by ‘utrip_id’, the first city and last city in the same trip were extracted to make a
dataframe.

And then the same paired cities are counted and the data frame has three columns,
‘first_city’, ‘fourth_city’, and ‘count’.

```python
testting=train_set
dfs=list(testting.groupby(testting['utrip_id']))
```

```python
newlist = []
for i in range(len(dfs)):
    x=dfs[i][1]['city_id']
    x = [x for x in x if isinstance(x, (int,float))]

    newlist.append(x)
```

```python
first_final_city=[]
for i in range(len(newlist)):
    first_final_city.append((newlist[i][0],newlist[i][-1]))
```

```python
#first_final_city
```

```python
number_of_pairs = collections.defaultdict(int)
for elem in first_final_city:
    number_of_pairs[elem] += 1
#number_of_pairs
```

```python
#number_of_pairs
```

```python
df = pd.DataFrame(number_of_pairs.items(), columns=['paired_cities','count'])
```

```python
df
```

```python
df[['first_city', 'fourth_city']] = pd.DataFrame(df['paired_cities'].tolist(), index=df.index)
df = df[['first_city', 'fourth_city', 'count']]
```

```python
df
```

```python
df.to_csv('training_first_fourth_city.csv', encoding='utf-8', index=False)
```
